```python
import pandas as pd
publi = pd.read_csv('http://faculty.marshall.usc.edu/gareth-james/ISL/Advertising.csv', index_col=0)
publi.head()
```

```python
print(publi.shape)
```

```python
x = publi[['newspaper','TV']]
y = publi['sales']
```

```python
x.shape
```

```python
y.shape
```

```python
import seaborn as sb
sb.pairplot(publi, x_vars=['newspaper','TV'], y_vars='sales', height=7, kind='reg')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

### importao do modelo de regresso linear

```python
from sklearn.linear_model import LinearRegression
reglin = LinearRegression()
reglin.fit(x_train, y_train)
```

```python
list(zip(['newspaper','TV'],reglin.coef_))
```

## Previso

```python
    y_previsao = reglin.predict(x_test)
    print(y_previsao)
```

### Avaliao da Performance
#### MAE (Mean absolute error)

```python
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(y_test,y_previsao))
```

#### MSE (Mean squared error)

```python
print(metrics.mean_squared_error(y_test, y_previsao))
```

#### RMSE (Root Mean squared error)

```python
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test, y_previsao)))
```
